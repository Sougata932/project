```python
calendar = pd.read_csv('calendar.csv')
dictionary = pd.read_csv('data_dictionary.csv')
inventory = pd.read_csv('inventory.csv')
products = pd.read_csv('products.csv')
sales = pd.read_csv('sales.csv')
stores = pd.read_csv('stores.csv')
```

```python
sales_products=pd.merge(sales,products, how='inner', on='Product_ID')
```

```python
sales_products
```

```python
sales_products_stores=pd.merge(sales_products, stores, how='inner', on='Store_ID')
```

```python
sales_products_stores.head()
```

```python
inventory.head()
```

```python
all=pd.merge(sales_products_stores,inventory, how='inner', on=['Store_ID', 'Product_ID'])
```

```python
all.head()
```

```python
all['Product_Name'].value_counts()
```

```python
plt.figure(figsize = (40,10))
sns.barplot(all['Product_Name'].value_counts(), palette='coolwarm', edgecolor='black')
plt.title("Distribution")
plt.xlabel('Product_Name')
plt.ylabel("count")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(all['Store_Location'].value_counts(), palette ='Set2', edgecolor='red')
plt.title("Distribution")
plt.xlabel("Store_Location")
plt.ylabel("count")
plt.show()
```

```python
all['Product_Category'].value_counts()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(all['Product_Category'].value_counts(), palette='Set1', edgecolor='black')
plt.title('distribution')
plt.xlabel("Products Category")
plt.ylabel("count")
plt.show()
```
